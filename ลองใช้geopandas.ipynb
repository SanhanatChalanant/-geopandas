```python
import matplotlib.pyplot as plt
cities = gpd.read_file('C:/Users/Admin/Downloads/New folder/Introduction to GIS Analysis with GeoPandas using Python/Shapefiles/belgian_cities.shp')
fig, ax = plt.subplots(1)
cities.plot(ax=ax, cmap = 'cool', column = 'NAME_4')
AOI.plot(ax=ax ,facecolor='yellow' )
```

```python
cities_in_AOI = gpd.overlay(cities, AOI, how = 'intersection')
```

```python
cities_in_AOI
```

```python
cities_in_AOI.plot(figsize = (10,10), cmap = 'winter',column = 'NAME_4')
```

```python
cities_in_AOI['Area(km2)'] = cities_in_AOI.area/1000000
```
